```python
import scanpy as scp
import matplotlib.pyplot as pyplot
import seaborn as sns
import numpy as np
import pandas as pd
import os
```

# Create h5ad files

Take the data from cellranger and convert it to a h5ad format to be processed by scanpy pipeline.

```python
datasets = [
    ("data","wildtype","rep1"),
    ("data","wildtype","rep2"),
    ("data","wildtype","rep3"),
    ("data","T_mutant","rep1"),
    ("data","T_mutant","rep2"),
    ("data","Eomes_mutant","rep1"),
    ("data","Eomes_mutant","rep2")
]
```

```python
for dataset in datasets:
    path_data = os.path.join(*dataset,"matrix.mtx")
    path_genes = os.path.join(*dataset,"features.tsv")
    path_h5ad = os.path.join(*dataset,"adata.h5ad")

    adata = scp.read_mtx(path_data).transpose()
    adata.var = pd.read_csv(path_genes, sep="\t", header=None, names=["gene_id", "gene_name", "gene_type"])
    adata.var_names = adata.var["gene_name"].values
    adata.var_names_make_unique()

    adata.write_h5ad(path_h5ad)
```
